```python
bronze_df = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .option("delimiter", ";")
    .csv("Files/sales_initial.csv")
)

display(bronze_df)
```

```python
bronze_df.printSchema()
```

```python
(
    bronze_df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("bronze_sales")
)
```
